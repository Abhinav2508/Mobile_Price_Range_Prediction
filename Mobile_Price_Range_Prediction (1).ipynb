## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.


#### üîπ Hypothesis 1:
**"The average RAM differs significantly across different mobile price ranges."**

- Type: One-Way ANOVA Test  
- Why: RAM is a continuous variable and Price Range is a categorical variable with 4 levels.  
- Goal: To check if the mean RAM varies significantly between price groups.

---

#### üîπ Hypothesis 2:
**"The proportion of phones that support 4G is not the same across different price ranges."**

- Type: Chi-Square Test of Independence  
- Why: Both `four_g` and `price_range` are categorical variables.  
- Goal: To check if 4G support is dependent on the price range.

---

#### üîπ Hypothesis 3:
**"Phones with a touch screen have significantly different battery power compared to those without a touch screen."**

- Type: Independent Samples t-test  
- Why: Battery power is numerical and `touch_screen` is binary (0 or 1).  
- Goal: To check if mean battery power differs based on touch screen availability.


### Hypothetical Statement - 1 : Does RAM significantly differ across price ranges?

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

#### 1Ô∏è‚É£ Null Hypothesis (H‚ÇÄ):
There is **no significant difference** in the average RAM across different price ranges.

#### 2Ô∏è‚É£ Alternate Hypothesis (H‚ÇÅ):
There **is a significant difference** in the average RAM across different price ranges.

---

#### 3Ô∏è‚É£ Perform an appropriate statistical test:
We will use **One-Way ANOVA** to compare the means of RAM across multiple groups (price ranges).


#### 2. Perform an appropriate statistical test.

```python
# Perform Statistical Test to obtain P-Value
from scipy.stats import f_oneway

# Group RAM values by price range
group0 = df[df['price_range'] == 0]['ram']
group1 = df[df['price_range'] == 1]['ram']
group2 = df[df['price_range'] == 2]['ram']
group3 = df[df['price_range'] == 3]['ram']

# Perform one-way ANOVA
stat, p = f_oneway(group0, group1, group2, group3)
print("ANOVA Test Statistic:", stat)
print("P-Value:", p)
```

#### ‚úÖ Which statistical test have you done to obtain the P-Value?
I used the **One-Way ANOVA** test.

---

#### ‚úÖ Why did you choose the specific statistical test?
- RAM is a **numerical** variable.
- Price Range is a **categorical** variable with **more than two categories** (0, 1, 2, 3).
- We are comparing the **mean RAM values** across multiple groups.

---

#### üìå Final Conclusion:
Since the **P-Value = 0.0** is **less than 0.05**, we **reject the null hypothesis**.  
‚úÖ This means that there is a **statistically significant difference** in RAM across different price ranges.

### Hypothetical Statement - 2 : The proportion of phones that support 4G is not the same across different price ranges

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

- **Null Hypothesis (H‚ÇÄ):** The proportion of phones supporting 4G is the same across all price ranges.
- **Alternate Hypothesis (H‚ÇÅ):** The proportion of phones supporting 4G is **not** the same across all price ranges.


#### 2. Perform an appropriate statistical test.

```python
# Perform Statistical Test to obtain P-Value
from scipy.stats import chi2_contingency
import pandas as pd

# Create a contingency table
contingency_table = pd.crosstab(df['price_range'], df['four_g'])

# Perform Chi-Square Test
chi2_stat, p_val, dof, expected = chi2_contingency(contingency_table)

print(f"Chi-Square Test Statistic: {chi2_stat}")
print(f"P-Value: {p_val}")
```

 **Why Chi-Square Test?**  
  The variables `price_range` (categorical) and `four_g` (categorical: Yes/No) are both categorical in nature.  
  Hence, a **Chi-Square Test of Independence** is suitable to test the association between them.

---

#### ‚úÖ Results:

- **Chi-Square Test Statistic:** `3.18`  
- **P-Value:** `0.3647`

---

#### üéØ Final Interpretation:

Since **P-Value = 0.3647 > 0.05**, we **fail to reject the null hypothesis**.

> ‚úÖ **Conclusion:** The proportion of phones that support 4G **does not significantly differ** across different price ranges.


### Hypothetical Statement - 3 : Phones with a touch screen have significantly different battery power compared to those without a touch screen

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

- **Null Hypothesis (H‚ÇÄ):** There is **no significant difference** in battery power between phones with a touch screen and those without.
- **Alternate Hypothesis (H‚ÇÅ):** There **is a significant difference** in battery power between phones with a touch screen and those without.


#### 2. Perform an appropriate statistical test.

```python
# Perform Statistical Test to obtain P-Value
from scipy.stats import ttest_ind

# Split battery power based on touch screen presence (1 = Yes, 0 = No)
touch_yes = df[df['touch_screen'] == 1]['battery_power']
touch_no = df[df['touch_screen'] == 0]['battery_power']

# Perform Independent Samples t-test
t_stat, p_val = ttest_ind(touch_yes, touch_no, equal_var=False)  # Welch's t-test

print("t-test Statistic:", t_stat)
print("P-Value:", p_val)
```

##### Which statistical test have you done to obtain P-Value?

Independent Samples t-test (Welch‚Äôs t-test)

##### Why did you choose the specific statistical test?

- Because we are comparing the means of a continuous variable (battery power) between two independent groups (phones with and without a touch screen).

- Welch‚Äôs t-test is used when the variances or sample sizes of the two groups may differ.
